```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
import matplotlib
from pylab import *
from numpy import genfromtxt

matplotlib.rc('font', family='Arial')
```

```python
my_data = genfromtxt('fluence.csv', delimiter=',')
set = [0]


Xlim= X = my_data[:,0]
Ylim= Y = my_data[:,1]
#error = my_data[:,2]
X = np.hsplit(X,1)
Y = np.hsplit(Y,1)
#error = np.hsplit(error,3)


#for i in set:
#    plt.plot(X[i],Y[i],'r')
#    errorbar(X[i], Y[i], yerr=error[i], fmt='-o')


#annotation
plt.plot(X[0],Y[0],'r')
plt.plot([25,25],[25,1600],color ='b',  linewidth=1.5, linestyle="--")
plt.scatter([25,],[1600,], 50, color ='b')

plt.annotate(r'Speed = 25mm/s, fluence = 1600J/cm$^2$',
             xy=(25, 1600),  xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=12,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([0,125],[320,320],color ='b',  linewidth=1.5, linestyle="--")

plt.scatter([125,],[320,], 50, color ='b')

plt.annotate(r'Speed = 125 mm/s, fluence = 320/cm$^2$',
             xy=(125, 320),  xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=12,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

        

# Set xy axis ticks
ylim(Ylim.min()-10,Ylim.max()*1.1)
xlim(Xlim.min()-20,Xlim.max()*1.1)
#yticks(np.linspace(0,ylim.max(),0,endpoint=True))


#title
#plt.title('Fluence variation')
plt.xlabel('Scan speed (mm/s)',fontsize=16)
plt.ylabel('Fluence (J/cm$^2$)',fontsize=16)

plt.show()

# Save figure using 72 dots per inch
#savefig("gaussian.png",dpi=300)

```

```python

# Create a new figure of size 8x6 points, using 80 dots per inch
figure(figsize=(10,4), dpi=300)

# Create a new subplot from a grid of 1x1
subplot(1,1,1)

fig1 = plt.subplot(121)

mean = 0
variance = 1
sigma = math.sqrt(variance)
gap = 0

#gap = 1.8

x = np.linspace(-4,10,100)
y = []


for i in range(0,10):
    y.append(mlab.normpdf(x,mean+i*gap,sigma)*-1)


sumy = []
for i in range (0,10):
    sumy_temp = 0
    
    for k in range (0,i):
        sumy_temp += y[k]
    sumy.append(sumy_temp)
    

    
    
for i in [1,2,3]:
    plt.plot(x,sumy[i])
    annotate(r'n=%s' % i,
             xy=((i-1)*gap/2, sumy[i].min()), xycoords='data',
             xytext=(-80, -40), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->"))

plt.plot(x,y[1])
# Scan gap
axvline(x=0.,color='k',ls='dashed')
axvline(x=gap,color='k',ls='dashed')
annotate ('', (0, 0.05), (gap, 0.05), arrowprops={'arrowstyle':'<->'})


# another figure
fig2 = plt.subplot(122)

mean = 0
variance = 1
sigma = math.sqrt(variance)
gap = 1.8


x = np.linspace(-4,10,100)
y2 = []


for i in range(0,10):
    y2.append(mlab.normpdf(x,mean+i*gap,sigma)*-1)


sumy2 = []
for i in range (0,10):
    sumy2_temp = 0
    
    for k in range (0,i):
        sumy2_temp += y2[k]
    sumy2.append(sumy2_temp)
    

    
    
for i in [1,2,3]:
    plt.plot(x,sumy2[i])
    annotate(r'n=%s' % i,
             xy=((i-1)*gap/2, sumy2[i].min()), xycoords='data',
             xytext=(-80, -40), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->"))

plt.plot(x,y2[1])
# Scan gap
axvline(x=0.,color='k',ls='dashed')
axvline(x=gap,color='k',ls='dashed')
annotate ('', (0, 0.05), (gap, 0.05), arrowprops={'arrowstyle':'<->'})




plt.show()



```

```python

# Create a new figure of size 8x6 points, using 80 dots per inch
figure(figsize=(8,6), dpi=80)

# Create a new subplot from a grid of 1x1
subplot(1,1,1)

fig1 = plt.subplot(111)



mean = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(-4,10,100)
y1 = mlab.normpdf(x,mean,sigma)*-1
y2 = mlab.normpdf(x,mean+0.5,sigma)*-1
y3 = mlab.normpdf(x,mean+1,sigma)*-1
y4 = mlab.normpdf(x,mean+1.5,sigma)*-1
y5 = mlab.normpdf(x,mean+2,sigma)*-1
y6 = mlab.normpdf(x,mean+2.5,sigma)*-1
y7 = mlab.normpdf(x,mean+3,sigma)*-1
y8 = mlab.normpdf(x,mean+3.5,sigma)*-1
y9 = mlab.normpdf(x,mean+4,sigma)*-1
y10 = mlab.normpdf(x,mean+4.5,sigma)*-1
y11 = mlab.normpdf(x,mean+5,sigma)*-1
y12 = mlab.normpdf(x,mean+5.5,sigma)*-1
y13 = mlab.normpdf(x,mean+6,sigma)*-1
y14 = mlab.normpdf(x,mean+6.5,sigma)*-1
y15 = mlab.normpdf(x,mean+7,sigma)*-1


Y3 = y1+y2+y3
Y5 = y1+y2+y3+y4+y5
Y7 = Y5+y6+y7
Y10 = Y7+y8+y9+y10
Y15 = Y10+ y11+y12+y13+y14+y15

plt.plot(x,y1,'r',x,Y3,'g',x,Y5,'c',x,Y15,'black',x,Y10,'')


# 1
annotate(r'n=1',
         xy=(0, y1.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))


# 3
annotate(r'n=3',
         xy=(0.5, Y3.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))


# 5
annotate(r'n=5',
         xy=(1, Y5.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))

# 10
annotate(r'n=10',
         xy=(3.5, Y10.min()+0.2), xycoords='data',
         xytext=(80, -40), textcoords='offset points',fontsize=16,
         arrowprops=dict(arrowstyle="->"))

# 15
annotate(r'n=15',
         xy=(6, Y15.min()+0.2), xycoords='data',
         xytext=(80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))

# Scan gap
axvline(x=0.,color='k',ls='dashed')
axvline(x=2.25,color='k',ls='dashed')
annotate ('', (0, -2.25), (2.25, -2.25), arrowprops={'arrowstyle':'<->'})





# Y axis
# Set y ticks
ylim(Y15.min()-0.5,0)
yticks(np.linspace(0,Y15.max(),0,endpoint=True))
plt.ylabel('depth (m)', fontsize = 20)

# X axis
plt.xlabel('width (m)', fontsize = 20)
plt.setp(fig1.get_xticklabels(), visible=False)


# Save figure using 72 dots per inch
savefig("gaussian.png",dpi=300)

plt.show()



```
