```python
# Figure one: single ablation depth and width


fig = plt.figure(figsize=(12,4), dpi=600)
ax = plt.subplot(122)


#figure(figsize=(8,4), dpi=300)
#subplot(1,2,1)

my_data = genfromtxt('holesize-paper.csv', delimiter=',')
#data
Xlim= X = my_data[:,0]
Wlim= W = my_data[:,1]
Dlim= D = my_data[:,3]

#error
Werr= my_data[:,2]
Derr= my_data[:,4]

#convert matrix to array
X = np.hsplit(X,1)
W = np.hsplit(W,1)
D = np.hsplit(D,1)
Werr = np.hsplit(Werr,1)
Derr = np.hsplit(Derr,1)


#figure 1
#draw scatter diagram -  width
#ax.scatter(X[0],W[0],c = 'k',label="Width",marker='o',s=30)
ax.errorbar(X[0],W[0], yerr=Werr[0], fmt='o', c = 'k',markersize =5,markerfacecolor='none' )

#fit width data
fitW = np.polyfit(X[0],W[0],2)
pW = np.poly1d(fitW)
ax.plot(X[0],pW(X[0]),'k--',alpha=2)




#Axis:  set x and y lim and ticks
ax.set_xlim(-0.5,Xlim.max()*1.1)
ax.set_ylim(20,60)

#ticks
ax.tick_params(labelsize = 20)


# Make the y-axis label and tick labels match the line color.
ax.set_xlabel('Waveform number',fontsize=28)
ax.set_ylabel('Width(μm)',fontsize=28)


#grid lines
ax.xaxis.grid(True) 
ax.yaxis.grid(True) 





#figure 2
ax2 = plt.subplot(121)

#draw Depth on another axis
#ax2.scatter(X[0],D[0],label="Depth",marker='o',s=30, facecolors='none', edgecolors='r' )
ax2.set_ylabel('Depth(μm)', color='k',fontsize=16,rotation=90,labelpad=10)
ax2.errorbar(X[0],D[0], yerr=Derr[0], fmt='o',c='k', markersize =5, markerfacecolor='none')

#fit depth data
fitD = np.polyfit(X[0],D[0],2)
pD = np.poly1d(fitD)
ax2.plot(X[0],pD(X[0]),'k--',alpha=1)


#Axis:  set x and y lim and ticks
ax2.set_xlim(-0.5,Xlim.max()*1.1)
ax2.set_ylim(0,4)


#ticks
ax2.tick_params(labelsize = 20)

# Make the y-axis label and tick labels match the line color.
ax2.set_xlabel('Waveform number',fontsize=28)
ax2.set_ylabel('Depth(μm)',fontsize=28)



#legend
#lines, labels = ax.get_legend_handles_labels()
#lines2, labels2 = ax2.get_legend_handles_labels()
#ax.legend(lines + lines2, labels + labels2, scatterpoints=1,fancybox=True, shadow=True, ncol=1, fontsize=14, loc=4)


#grid lines
ax2.xaxis.grid(True) 
ax2.yaxis.grid(True) 



#setting title
#plt.suptitle('single ablation',fontsize= 30)


#layout
plt.tight_layout(pad=0.1, w_pad=5, h_pad=5)




# Save figure using 72 dots per inch
#savefig("holesize_paper.png",dpi=300,bbox_inches='tight')
#fig.savefig('samplefigure', bbox_extra_artists=(lgd,), bbox_inches='tight')

plt.show()
```

```python
# Figure two: channel machining


fig = plt.figure(figsize=(8,4), dpi=300)
ax = plt.subplot(111)


#figure(figsize=(8,4), dpi=300)
#subplot(1,2,1)

my_data = genfromtxt('channel_size.csv', delimiter=',')
#data
Xlim= X = my_data[:,0]
Wlim= W = my_data[:,1]
Dlim= D = my_data[:,3]
#error
Werr= my_data[:,2]
Derr= my_data[:,4]

#convert matrix to array
X = np.hsplit(X,1)
W = np.hsplit(W,1)
D = np.hsplit(D,1)
Werr = np.hsplit(Werr,1)
Derr = np.hsplit(Derr,1)


#draw scatter diagram -  width
ax.scatter(X[0],W[0],c = 'b',label="Width",marker='o',s=30,linewidths=1)
ax.errorbar(X[0],W[0], yerr=Werr[0], fmt='o', c = 'b')

#fit width data
fitW = np.polyfit(X[0],W[0],1)
pW = np.poly1d(fitW)
ax.plot(X[0],pW(X[0]),'b--',alpha=2)
ax.tick_params(axis='y', colors='b')


#draw Depth on another axis
ax2 = ax.twinx()
ax2.scatter(X[0],D[0],c = 'r',label="Depth",marker='o',s=30)
ax2.set_ylabel('Depth (μm)', color='r',fontsize=16,rotation=90,labelpad=10)
ax2.errorbar(X[0],D[0], yerr=Derr[0], fmt='o', c = 'r')


#fit depth data
fitD = np.polyfit(X[0],D[0],1)
pD = np.poly1d(fitD)
ax2.plot(X[0],pD(X[0]),'r--',alpha=1)


#Axis:  set x and y lim and ticks
ax.set_xlim(0,Xlim.max()*1.1)
ax.set_ylim(0,60)
ax2.set_ylim(0,4)

# Make the y-axis label and tick labels match the line color.
ax.set_xlabel('Peak power intensity (MW/mm$^2$)',fontsize=16)
ax.set_ylabel('Width (μm)', color='b',fontsize=16)
    

    

#legend
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, scatterpoints=1,fancybox=True, shadow=True, ncol=1, fontsize=14, loc=4)


#grid lines
ax.xaxis.grid(True) 

#layout
#plt.tight_layout(pad=0.1, w_pad=2, h_pad=.1)


# Save figure using 72 dots per inch
#savefig("holesize_paper.png",dpi=300,bbox_inches='tight')
#fig.savefig('samplefigure', bbox_extra_artists=(lgd,), bbox_inches='tight')

plt.show()
```
