```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
import matplotlib
from pylab import *
from numpy import genfromtxt

matplotlib.rc('font', family='Arial')
```

```python
figure(figsize=(12,4), dpi=300)
subplot(1,2,1)

my_data = genfromtxt('holesize-paper.csv', delimiter=',')
set = [0]

Xlim= X = my_data[:,0]
Wlim= W = my_data[:,1]
Dlim= D = my_data[:,2]
Alim= A = my_data[:,3]

#error = my_data[:,2]
X = np.hsplit(X,1)
W = np.hsplit(W,1)
D = np.hsplit(D,1)
A = np.hsplit(A,1)
#error = np.hsplit(error,3)

#fig, ax1 = plt.subplots()
#fig, (ax1,ax3) = plt.subplots(2, sharex=True)


plt.scatter(X[0],W[0],c = 'b',label="Measured width",marker='o',s=30)
#ax1.set_xlabel('Peak power intensity (MW/mm$^2$)',fontsize=16)
# Make the y-axis label and tick labels match the line color.
plt.ylabel('Width (um)', color='b',fontsize=16)
#fit
fitW = np.polyfit(X[0],W[0],1)
pW = np.poly1d(fitW)
plt.plot(X[0],pW(X[0]),'b--',alpha=1)
plt.tick_params(axis='y', colors='b')

ylim(0,Wlim.max()*1.2)

#legend
legend(scatterpoints=1,fancybox=True, shadow=True, ncol=2, fontsize=14, loc=2)



#Depth
ax2 = plt.twinx()
ax2.scatter(X[0],D[0],c = 'r',label="Measured depth",marker='o',s=30)
ax2.set_ylabel('Depth (um)', color='r',fontsize=16,rotation=90,labelpad=20)
#fit
fitD = np.polyfit(X[0],D[0],1)
pD = np.poly1d(fitD)
ax2.plot(X[0],pD(X[0]),'r--',alpha=1)

ylim(0,Dlim.max()*1.1)
for tl in ax2.get_yticklabels():
    tl.set_color('r')

    
#legend
plt.grid(True)
legend(scatterpoints=1,fancybox=True, shadow=True, ncol=2, fontsize=14, loc=4)





    
subplot(1,2,2)

#Aspect ratio
plt.scatter(X[0],A[0],c = 'k',label="Measured aspect ratio",marker='o',s=30)
#fit
fitA = np.polyfit(X[0],A[0],1)
pA = np.poly1d(fitA)
plt.plot(X[0],pA(X[0]),'k--',alpha=1,label="Modelled aspect ratio")

plt.xlabel('Peak power intensity (MW/mm$^2$)',fontsize=16)
plt.ylabel('Depth-to-width aspect ratio', color='k',fontsize=16)
ylim(0,Alim.max()*1.2)
xlim(20,Xlim.max()*1.1)


#legend
plt.grid(True)
legend(scatterpoints=1,fancybox=True, shadow=True, ncol=1, fontsize=14, loc=2)





#Width
#plt.scatter(X[0],W[0],c = 'r',label="Width",marker='o',s=50)
#fitW = np.polyfit(X[0],W[0],1)
#pW = np.poly1d(fitW)
#plt.plot(X[0],pW(X[0]),'r--',alpha=1,label="Trend width")



# Set xy axis ticks
#ylim(0,Wlim.max()*1.1)
#xlim(Xlim.min(),Xlim.max()*1.1)
#yticks(np.linspace(0,1800,10,endpoint=True))


#title
#plt.title('Fluence variation')
#plt.xlabel('Fluence (J/cm$^2$)',fontsize=16)
#plt.ylabel('Measured dimension (μm)',fontsize=16)


#legend
grid(True)


#layout
plt.tight_layout(pad=0.2, w_pad=0.7, h_pad=.5)


# Save figure using 72 dots per inch
#savefig("holesize.png",dpi=300,bbox_inches='tight')
#fig.savefig('samplefigure', bbox_extra_artists=(lgd,), bbox_inches='tight')

plt.show()
```

```python

# Create a new figure of size 8x6 points, using 80 dots per inch
figure(figsize=(8,6), dpi=80)

# Create a new subplot from a grid of 1x1
subplot(1,1,1)

fig1 = plt.subplot(111)



mean = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(-4,10,100)
y1 = mlab.normpdf(x,mean,sigma)*-1
y2 = mlab.normpdf(x,mean+0.5,sigma)*-1
y3 = mlab.normpdf(x,mean+1,sigma)*-1
y4 = mlab.normpdf(x,mean+1.5,sigma)*-1
y5 = mlab.normpdf(x,mean+2,sigma)*-1
y6 = mlab.normpdf(x,mean+2.5,sigma)*-1
y7 = mlab.normpdf(x,mean+3,sigma)*-1
y8 = mlab.normpdf(x,mean+3.5,sigma)*-1
y9 = mlab.normpdf(x,mean+4,sigma)*-1
y10 = mlab.normpdf(x,mean+4.5,sigma)*-1
y11 = mlab.normpdf(x,mean+5,sigma)*-1
y12 = mlab.normpdf(x,mean+5.5,sigma)*-1
y13 = mlab.normpdf(x,mean+6,sigma)*-1
y14 = mlab.normpdf(x,mean+6.5,sigma)*-1
y15 = mlab.normpdf(x,mean+7,sigma)*-1


Y3 = y1+y2+y3
Y5 = y1+y2+y3+y4+y5
Y7 = Y5+y6+y7
Y10 = Y7+y8+y9+y10
Y15 = Y10+ y11+y12+y13+y14+y15

plt.plot(x,y1,'r',x,Y3,'g',x,Y5,'c',x,Y15,'black',x,Y10,'')


# 1
annotate(r'n=1',
         xy=(0, y1.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))


# 3
annotate(r'n=3',
         xy=(0.5, Y3.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))


# 5
annotate(r'n=5',
         xy=(1, Y5.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))

# 10
annotate(r'n=10',
         xy=(2.25, Y10.min()), xycoords='data',
         xytext=(80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))

# 15
annotate(r'n=15',
         xy=(4, Y15.min()), xycoords='data',
         xytext=(80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))

# Scan gap
axvline(x=0.,color='k',ls='dashed')
axvline(x=2.25,color='k',ls='dashed')
annotate ('', (0, -2.25), (2.25, -2.25), arrowprops={'arrowstyle':'<->'})





# Y axis
# Set y ticks
ylim(Y15.min()-0.5,0)
yticks(np.linspace(0,Y15.max(),0,endpoint=True))
plt.ylabel('depth', fontsize = 20)

# X axis
plt.xlabel('width', fontsize = 20)
plt.setp(fig1.get_xticklabels(), visible=False)


# Save figure using 72 dots per inch
savefig("gaussian.png",dpi=300)

plt.show()



```

```python
# Create a new figure of size 8x6 points, using 80 dots per inch
figure(figsize=(8,6), dpi=80)

# Create a new subplot from a grid of 1x1
subplot(1,1,1)

fig2 = plt.subplot(111)



mean = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(-4,10,100)
y1 = mlab.normpdf(x,mean,sigma)*-1
y2 = mlab.normpdf(x,mean+4,sigma)*-1
y3 = mlab.normpdf(x,mean+8,sigma)*-1
y4 = mlab.normpdf(x,mean+12,sigma)*-1
y5 = mlab.normpdf(x,mean+16,sigma)*-1


Y2 = y1+y2
Y3 = y1+y2+y3


plt.plot(x,Y3,'b',x,Y2,'g',x,y1,'r')


# 1
annotate(r'n=1',
         xy=(0, y1.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))


# 3
annotate(r'n=2',
         xy=(4, Y2.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))


# 5
annotate(r'n=3',
         xy=(8, Y3.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))


# Scan gap
axvline(x=0.,color='k',ls='dashed')
axvline(x=4.,color='k',ls='dashed')
annotate ('', (0, -1.5), (4, -1.5), arrowprops={'arrowstyle':'<->'})




# Y axis
# Set y ticks
ylim(-1.9997-0.5,0)
yticks(np.linspace(0,Y15.max(),0,endpoint=True))
plt.ylabel('depth', fontsize = 20)

# X axis
plt.xlabel('width', fontsize = 20)
plt.setp(fig2.get_xticklabels(), visible=False)



# Save figure using 72 dots per inch
savefig("gaussian_2.png",dpi=300)



plt.show()



```
