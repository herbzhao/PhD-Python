```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
import matplotlib
from pylab import *
from numpy import genfromtxt

matplotlib.rc('font', family='Arial')
```

```python
figure(figsize=(8,8), dpi=300)
subplot(2,1,1)

my_data = genfromtxt('width.csv', delimiter=',')
set = [0]

Xlim= X = my_data[:,0]
Wlim= W = my_data[:,1]
Dlim= D = my_data[:,2]

#error = my_data[:,2]
X = np.hsplit(X,4)
W = np.hsplit(W,4)
D = np.hsplit(D,4)
#error = np.hsplit(error,3)


plt.plot(X[0],W[0],'r.-',label="300mm/s, 40W, 0, d=200m,unblown",linewidth=1.5)
plt.plot(X[2],W[2],'b.-',label="300mm/s, 40W, 0, d=30um",linewidth=1.5)
plt.plot(X[1],W[1],'g.-',label="300mm/s, 40W, 0, d=200um",linewidth=1.5)
plt.plot(X[3],W[3],'m.-',label="300mm/s, 40W, 90, d=30um",linewidth=1.5)

# Set xy axis ticks
ylim(Wlim.min()-200,Wlim.max()*1.1)
xlim(Xlim.min()-200,Xlim.max()*1.1)
#yticks(np.linspace(0,ylim.max(),0,endpoint=True))


#title
#plt.title('Fluence variation')
#plt.xlabel('Designed width ($\mu m$)',fontsize=16)

plt.ylabel('Actual width (m)',fontsize=16)
plt.grid(True)

subplot(2,1,2)

plt.plot(X[0],D[0],'r.-',label="300 mm/s, 40 W, 0, D=200 m, unblown",linewidth=1.5)
plt.plot(X[2],D[2],'b.-',label="300 mm/s, 40 W, 0, D=30 m",linewidth=1.5)
plt.plot(X[1],D[1],'g.-',label="300 mm/s, 40 W, 0, D=200 m",linewidth=1.5)
plt.plot(X[3],D[3],'m.-',label="300 mm/s, 40 W, 90, D=30 m",linewidth=1.5)



# Set xy axis ticks
ylim(Dlim.min()-200,Dlim.max()*1.1)
xlim(Xlim.min()-200,Xlim.max()*1.1)
yticks(np.linspace(0,1800,10,endpoint=True))


#title
#plt.title('Fluence variation')
plt.xlabel('Designed width (m)',fontsize=16)
plt.ylabel('Actual depth (m)',fontsize=16)


#legend
plt.grid(True)

legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
          fancybox=True, shadow=True, ncol=2, fontsize=14)

#layout
plt.tight_layout(pad=0.2, w_pad=0.7, h_pad=.5)

#plt.show()

# Save figure using 72 dots per inch
savefig("width and depth.png",dpi=300,bbox_inches='tight')
#fig.savefig('samplefigure', bbox_extra_artists=(lgd,), bbox_inches='tight')


```

```python

# Create a new figure of size 8x6 points, using 80 dots per inch
figure(figsize=(8,6), dpi=80)

# Create a new subplot from a grid of 1x1
subplot(1,1,1)

fig1 = plt.subplot(111)



mean = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(-4,10,100)
y1 = mlab.normpdf(x,mean,sigma)*-1
y2 = mlab.normpdf(x,mean+0.5,sigma)*-1
y3 = mlab.normpdf(x,mean+1,sigma)*-1
y4 = mlab.normpdf(x,mean+1.5,sigma)*-1
y5 = mlab.normpdf(x,mean+2,sigma)*-1
y6 = mlab.normpdf(x,mean+2.5,sigma)*-1
y7 = mlab.normpdf(x,mean+3,sigma)*-1
y8 = mlab.normpdf(x,mean+3.5,sigma)*-1
y9 = mlab.normpdf(x,mean+4,sigma)*-1
y10 = mlab.normpdf(x,mean+4.5,sigma)*-1
y11 = mlab.normpdf(x,mean+5,sigma)*-1
y12 = mlab.normpdf(x,mean+5.5,sigma)*-1
y13 = mlab.normpdf(x,mean+6,sigma)*-1
y14 = mlab.normpdf(x,mean+6.5,sigma)*-1
y15 = mlab.normpdf(x,mean+7,sigma)*-1


Y3 = y1+y2+y3
Y5 = y1+y2+y3+y4+y5
Y7 = Y5+y6+y7
Y10 = Y7+y8+y9+y10
Y15 = Y10+ y11+y12+y13+y14+y15

plt.plot(x,y1,'r',x,Y3,'g',x,Y5,'c',x,Y15,'black',x,Y10,'')


# 1
annotate(r'n=1',
         xy=(0, y1.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))


# 3
annotate(r'n=3',
         xy=(0.5, Y3.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))


# 5
annotate(r'n=5',
         xy=(1, Y5.min()), xycoords='data',
         xytext=(-80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))

# 10
annotate(r'n=10',
         xy=(2.25, Y10.min()), xycoords='data',
         xytext=(80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))

# 15
annotate(r'n=15',
         xy=(4, Y15.min()), xycoords='data',
         xytext=(80, -40), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->"))

# Scan gap
axvline(x=0.,color='k',ls='dashed')
axvline(x=2.25,color='k',ls='dashed')
annotate ('', (0, -2.25), (2.25, -2.25), arrowprops={'arrowstyle':'<->'})





# Y axis
# Set y ticks
ylim(Y15.min()-0.5,0)
yticks(np.linspace(0,Y15.max(),0,endpoint=True))
plt.ylabel('depth', fontsize = 20)

# X axis
plt.xlabel('width', fontsize = 20)
plt.setp(fig1.get_xticklabels(), visible=False)


# Save figure using 72 dots per inch
savefig("gaussian.png",dpi=300)

plt.show()



```
