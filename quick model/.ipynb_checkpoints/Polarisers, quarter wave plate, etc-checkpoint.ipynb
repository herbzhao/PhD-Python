```python
# plotting
class plotly_plot(object):
    def __init__(self):
        pass
            
    def plotly_plotting(self, data):
        self.traces = {}
        for i in data:
            self.traces[i] = go.Scatter3d(
                #self.traces[i] = go.Scattergl( # performs better at larger dataset
                    x = data[i][0],
                    y = data[i][1],
                    z = data[i][2],
                    name = 'simulated wave',
                    mode = self.plot_type,
                    #text='RH {}'.format(self.peak_data[sample_number]['humidity'][scan_number]), # hover text
                    line = dict(
                        #dash = 'dash', 
                        #color = '',
                        ),
                     marker = dict(
                        size = 10,
                        #color = 'rgba(255, 0, 0, .8)',
                        )
                    )

        
        
        # modify layout with dict()
        layout = dict(
            scene = dict(
            xaxis= dict( 
                range = self.x_lim, # move the x_axis to start from 0.
                title= self.x_label,
                titlefont=dict(size=30*self.font_size_factor, family = 'Times New Roman'),
                ticklen= 0,
                tickfont=dict(size=20*self.font_size_factor, family = 'Times New Roman'),
                showline=True,
                zeroline= True,                
                gridwidth= 1,
                showgrid=True,
                ),
            yaxis = dict(
                range = self.y_lim,
                title = self.y_label,
                titlefont=dict(size=30*self.font_size_factor, family = 'Times New Roman'),
                ticklen= 0,
                tickfont=dict(size=20*self.font_size_factor, family = 'Times New Roman'),
                showline = True,
                zeroline = True,
                showgrid = True,
                gridwidth= 1,
                ),),
            showlegend = self.show_legend,
            legend = dict(
                font=dict(size=22*self.font_size_factor, family = 'Times New Roman'),
                #yanchor='middle',
                #xanchor='middle'),
                x = self.legend_position[0],
                y = self.legend_position[1],
                # a box for legend
                bgcolor='white',
                #bordercolor='black',
                borderwidth=2,
                ), 
            margin = go.Margin(
                l=75*self.font_size_factor,
                #r=50,
                #b=100,
                #t=100,
                #pad=5
                ),
            # transparent graph
            paper_bgcolor='rgba(0,0,0,0)',
            plot_bgcolor='rgba(0,0,0,0)'
            )

        
        # create data -- a list contains all the traces
        data = []
        for i in self.traces:
            data.append(self.traces[i])
        # create figure with data and layout
        fig = dict(data=data, layout=layout)
        # set export static image
        
        # generate the plot and save as svg
        py.iplot(fig)
        '''py.iplot(fig,  filename='csv_plot', 
                     image_width = self.image_size[0],
                     image_height = self.image_size[1], 
                     image='svg',)
        '''
        


    
    
    
```

```python
class optical_simulation_components(object):

    def __init__(self):
        # every polariser added, extend the propogation 
        self.propogation_number = 0
        # length before reaching another optical component
        self.propogation_length = 1
        self.propogation_unit = np.linspace(0, self.propogation_length, 1000)
        # operator to extend the propogation
        self.propogation_extention = np.array([[0],[0],[self.propogation_length]])
        # 0 degrees linear polarised light
        self.wave = {}
        self.wave[self.propogation_number] = np.array([np.sin(self.propogation_unit*2*np.pi), self.propogation_unit*0, self.propogation_unit])

    def linear_polariser_matrix(self, polarisation_angle):
        # intensity at polarisation direction
        linear_polariser_operator = np.array([
            [np.cos(polarisation_angle), np.sin(polarisation_angle), 0],
            [0, 0, 0],
            [0, 0, 1]])
        # polarised wave on polarisation direction
        linear_polarised_wave = linear_polariser_operator.dot(self.wave[self.propogation_number])
        
        # projection operator
        linear_polariser_operator_2 = np.array([
            [np.cos(polarisation_angle), np.sin(polarisation_angle),0],
            [np.sin(polarisation_angle), np.cos(polarisation_angle),0],
            [0,0,1]])
        # projection onto x and y axis
        self.propogation_number += 1
        self.wave[self.propogation_number] = linear_polariser_operator_2.dot(linear_polarised_wave) + self.propogation_extention*self.propogation_number
        
    def retardation_plate(self, slow_axis_angle, phase_difference):
        # intensity at polarisation direction
        retardation_operator = np.array([
            [np.cos(slow_axis_angle), np.sin(slow_axis_angle),0],
            [np.sin(slow_axis_angle), np.cos(slow_axis_angle),0],
            [0,0,1]])
        
        retardation_wave = retardation_operator.dot(self.wave[self.propogation_number])
        # roll the array 
        roll_factor = phase_difference/np.pi/2%1
        retardation_wave = np.stack(
            (np.roll(retardation_wave[0], int(roll_factor*(retardation_wave[0].shape[0]))+1),
             retardation_wave[1],
             retardation_wave[2],)
            )

        # projection onto x and y axis
        self.propogation_number += 1
        self.wave[self.propogation_number] = retardation_operator.dot(retardation_wave)+ self.propogation_extention
        
        
```

```python


#This code simulates the efffect of different diffraction index
#in different polarised direction on polarised light

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure(figsize=(24,24))
ax = fig.gca(projection='3d')

'the parameters that require changes'
#n is the number of periods (light path in bifriengence material)
n = 3
#phase lag caused by n1-n2
phase_diff = np.pi/2    



theta = np.linspace( 0,  2*n*np.pi, 100)
#indicate the rotating axis
x0 = np.linspace(0, 0, 100)
y0 = np.linspace(0, 2*n*np.pi, 100)
z0 = np.linspace(0, 0, 100)
ax.plot(x0, y0 ,z0, label='central', c= 'k')

#component light 1
x1 = np.linspace(0, 0, 100)
y1 = np.linspace(0, 2*n*np.pi, 100)
z1 = np.sin(theta)
ax.plot(x1, y1 ,z1, label='z-axis')

#component light 2



x2 = np.sin(theta+phase_diff)  
#this creates the phase difference due to liquid crystal
y2 = np.linspace(0, 2*n*np.pi, 100)
z2 = np.linspace(0, 0, 100)
ax.plot(x2,y2, z2, label='x-axis')

#combined circular polarised light
x3 = (x1+x2)
y3 = (y1+y2)/2
z3 = z1+z2
ax.plot(x3,y3, z3, label='combined')

    
ax.legend()

plt.show()
```
